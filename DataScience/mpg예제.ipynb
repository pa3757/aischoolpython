### 목표 설정
- 실제 데이터인 mpg데이터의 특징을 파악하고 분석해보자
- mpg(miles per gallon)데이터는 미국 환경 보호국에서 공개한 데이터
- 1999 ~ 2008년까지 출시된 자동차 234종에 대한 정보

```python
import pandas as pd

mpg = pd.read_csv('data/mpg.csv')
mpg
```

```python
# 앞의 행 3개를 출력해서 데이터를 확인해보자
mpg.head(3)
```

```python
# 뒤의 행 3개를 출력해서 데이터를 확인해보자
mpg.tail(3)
```

```python
# 데이터가 몇행 몇열로 구성되어있는지 확인
mpg.shape
```

```python
# 전체적인 데이터의 속성 정보 확인하기
# 데이터프레임의 요약정보 확인하기
mpg.info()
```

```python
# 요약 통계량 살펴보기
# describe() : 설명하다, 묘사하다 -> 데이터의 요약 통계량을 보여준다
mpg.describe()
```

```python
mpg
```

```python
# audi에서 생산한 자동차 중 hwy가 1~5위에 해당하는 자동차 데이터 출력하기
# 자동차 제조사 컬럼에서 audi만 뽑아오기(불리언 인덱싱/query)
# audi만 뽑은 데이터에서 hwy 값을 기준으로 내림차순 정렬하기
# 정렬한 데이터를 가지고 상위 5개의 데이터를 확인하기
mpg[mpg['manufacturer'] == 'audi'].sort_values(by='hwy', ascending=False).head(5)
```

```python
# 차량 카테고리별 도시 연비 평균
# groupby()
# 차량 카테고리 개수 확인
mpg['category'].value_counts()
```

```python
# 필요한 컬럼들만 묶어주기 groupby()
mpg[['category','cty']].groupby('category').mean()
```

```python
# 다른 방법으로 접근 - agg
# agg : 다중 집계함수 , 여러개의 값을 받아서 일괄적으로 연산진행
cty_mean = mpg.groupby('category').agg({'cty':'mean'})
```

```python
cty_mean
```

```python
# 위에서 조사한 연비 평균 데이터를 이용해서 어떤 차종의 연비가 좋은지 내림차순으로 정렬
cty_mean.sort_values(by='cty', ascending=False)
```

```python
# hwy 평균이 가장 높은 회사 세곳
mpg[['manufacturer','hwy']].groupby('manufacturer').mean().sort_values(by='hwy', ascending=False).head(3)
```

```python
# 회사별 compact차종 수 내림차순 정렬
mpg[mpg['category']=='compact']['manufacturer'].value_counts().sort_values(ascending=False)
```

```python
# query함수 사용하기
mpg.query('category == "compact"')['manufacturer'].value_counts()
```
