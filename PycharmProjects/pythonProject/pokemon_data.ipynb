```python
import pandas as pd
requests = __import__('requests')

url = 'https://pokeapi.co/api/v2/pokemon/'
koreannameurl = 'https://pokeapi.co/api/v2/pokemon-species'

res = requests.get(url)
res2 = requests.get(koreannameurl+ '/1')

pokemon = res.json()
pokemon2 = res2.json()
```

```python
pokemon_list = []
for i in range(1, 151):
    res = requests.get(url + str(i))
    pokemon = res.json()
    res2 = requests.get(koreannameurl + '/' + str(i))
    pokemon2 = res2.json()
    pokemon_list.append({
        'poke_num' : pokemon['id'],
        'poke_name': pokemon2['names'][2]['name'],
        'poke_type': [t['type']['name'] for t in pokemon['types']],
        #'poke_img': pokemon['sprites']['front_default'],
        'poke_img' : pokemon['sprites']['other']['showdown']['front_default']
    })
```

```python
# dataframe으로 변환
df = pd.DataFrame(pokemon_list)
df
```

```python
for i in range(0,150):
    df['types'][i] = df['types'][i][0]
df
```

```python
df.to_csv('pokemon_data.csv', index=False)
```

```python
# csv 파일 utf-8로 저장
df.to_csv('pokemon_data.csv', index=False, encoding='utf-8')
```

```python
df = pd.read_csv('pokemon_data.csv')
df
```

```python
# poke_type 을 mysql에 삽입할 수 있게 json으로 변환
df['poke_type'] = df['poke_type'].apply(lambda x: str(x).replace('[','').replace(']','').replace("'",''))
df
```

```python
df.to_csv('pokemon_data.csv', index=False, encoding='utf-8')
```

```python
# csv 파일 문장 앞뒤로 () 추가 후 txt 파일로 저장
# ex) (1,이상해씨,"grass, poison",https://raw.githubusercontent.com/PokeAPI/sprites/master/sprites/pokemon/other/showdown/1.gif)
df['poke_type'] = '(' + df['poke_num'].astype(str) + ',' + df['poke_name'] + ',"' + df['poke_type'] + '",' + df['poke_img'] + ')'
df.to_csv('pokemon_data.txt', index=False, header=False, encoding='utf-8')
```
