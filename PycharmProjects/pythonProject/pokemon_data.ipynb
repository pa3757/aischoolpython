```python
import pandas as pd
requests = __import__('requests')

url = 'https://pokeapi.co/api/v2/pokemon/'
koreannameurl = 'https://pokeapi.co/api/v2/pokemon-species'

res = requests.get(url)
res2 = requests.get(koreannameurl+ '/1')

pokemon = res.json()
pokemon2 = res2.json()
```

```python
pokemon_list = []
for i in range(1, 151):
    res = requests.get(url + str(i))
    pokemon = res.json()
    res2 = requests.get(koreannameurl + '/' + str(i))
    pokemon2 = res2.json()
    pokemon_list.append({
        'id' : pokemon['id'],
        'name': pokemon['name'],
        'koreanname': pokemon2['names'][2]['name'],
        'height': pokemon['height'],
        'weight': pokemon['weight'],
        'types': [t['type']['name'] for t in pokemon['types']],
        'img-url': pokemon['sprites']['front_default'],
        'gif-url' : pokemon['sprites']['other']['showdown']['front_default']
    })
```

```python
# dataframe으로 변환
df = pd.DataFrame(pokemon_list)
df
```

```python
for i in range(0,150):
    df['types'][i] = df['types'][i][0]
df
```

```python
df.to_csv('pokemon_data.csv', index=False)
```
